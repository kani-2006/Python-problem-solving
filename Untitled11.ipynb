```python
#Print Numbers from 1 to N
n=int(input())
for i in range(1,6):
  print(i,end=" ")
```

```python
#Print Even Numbers up to N
n=int(input())
for i in range(1,11):
  if i%2==0:
    print(i,end=" ")
```

```python
#Sum of First N Natural Numbers
n=int(input())
sum=0
for i in range(1,n+1):
  sum+=i
print(sum)
```

```python
#Multiplication Table
n=int(input())
for i in range(1,11):
  print(n,"*",i,"=",n*i)
```

```python
#Factorial of a Number
n=int(input())
fact=1
for i in range(1,n+1):
  fact*=i
print(fact)
```

```python
#Count Digits in a Number
n=int(input())
count=0
for i in str(n):
  count+=1
print(count)
```

```python
#Reverse a Number
n=int(input())
print(str(n)[::-1])
```

```python
#Sum of Digits
n=int(input())
sum=0
for i in str(n):
  sum += int(i)
print(sum)
```

```python
#Print Squares from 1 to N
n=int(input())
for i in range(1,n+1):
  print(i*i,end=" ")
```

```python
#Print Odd Numbers in a Range
n,l=map(int,input().split())
for i in range(n,l):
  if i%2!=0:
    print(i,end=" ")
```

```python
#Number Divisors
n=int(input())
for i in range(1,n+1):
  if n%i==0:
    print(i,end=" ")
```
